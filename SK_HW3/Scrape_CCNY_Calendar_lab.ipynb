```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# fetch webpage
url = "https://www.ccny.cuny.edu/registrar/fall"
response = requests.get(url)

# check status
print(response.status_code)
```

```python
# parse html
soup = BeautifulSoup(response.text, "html.parser")

# find table
table = soup.find("table")

# get rows
rows = table.find_all("tr")

print(len(rows))
```

```python
# store data
data = []

# skip header row
for row in rows[1:]:
    cols = row.find_all("td")
    
    date = cols[0].get_text(strip=True)
    dow = cols[1].get_text(strip=True)
    text = cols[2].get_text(" ", strip=True)
    
    data.append([date, dow, text])

print(data[:5])
```

```python
# create dataframe
df = pd.DataFrame(data, columns=["date", "dow", "text"])

df.head()
```

```python
# clean date ranges
def clean_date(date_str):
    
    # take first date
    date_str = date_str.split("-")[0].strip()
    
    # convert to datetime
    return pd.to_datetime(date_str + " 2021")

# apply function
df["date"] = df["date"].apply(clean_date)

df.head()
```

```python
# set date index
df = df.set_index("date")

# sort index
df = df.sort_index()

df.head()
```
