In this workspace, you'll scrape the novel Moby Dick from the website [Project Gutenberg](https://www.gutenberg.org/) (which contains a large corpus of books) using the Python `requests` package. You'll extract words from this web data using `BeautifulSoup` before analyzing the distribution of words using the Natural Language ToolKit (`nltk`) and `Counter`.

The Data Science pipeline you'll build in this workspace can be used to visualize the word frequency distributions of any novel you can find on Project Gutenberg.

```python
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import nltk

nltk.download('punkt')
nltk.download('stopwords')

source_url = 'https://s3.amazonaws.com/assets.datacamp.com/production/project_147/datasets/2701-h.htm'
response = requests.get(source_url)
response.encoding = 'utf-8'

raw_content = response.text
print(raw_content[:2000])

soup_data = BeautifulSoup(raw_content, "html.parser")
full_text = soup_data.get_text()

raw_tokens = word_tokenize(full_text)
cleaned_tokens = [t.lower() for t in raw_tokens if t.isalpha()]

print(cleaned_tokens[:8])

english_stops = set(stopwords.words('english'))
print(list(english_stops)[:8])

filtered_list = [w for w in cleaned_tokens if w not in english_stops]

print(filtered_list[:5])

frequency_map = Counter(filtered_list)
most_frequent = frequency_map.most_common(10)

print(most_frequent)
```
