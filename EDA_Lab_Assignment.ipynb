```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
The dataset contains 891 rows and 12 columns with both numerical and categorical variables. The Age column has some missing values, and the Cabin column has many missing values, while Fare shows high variability.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# HISTOGRAMS (NUMERICAL ONLY)
df[['Age','Fare','SibSp','Parch']].hist(figsize=(10,8))
plt.show()

# BAR CHARTS (CATEGORICAL)
sns.countplot(x='Sex', data=df)
plt.title("Sex Distribution")
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()

sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution")
plt.show()

# BOXPLOT (ALL NUMERICAL TOGETHER)
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['Age','Fare','SibSp','Parch']])
plt.title("Boxplot of Numerical Variables")
plt.show()

# PAIRPLOT (NUMERICAL RELATIONSHIPS)
sns.pairplot(df[['Age','Fare','SibSp','Parch']])
plt.show()

# CATEGORICAL vs CATEGORICAL
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar')
plt.title("Sex vs Survival")
plt.show()

# CATEGORICAL vs NUMERICAL
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()

# COLOR SCATTER
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```

```python
The summary statistics show that the minimum and maximum values vary significantly across variables, with Age ranging from approximately 0 to 80 and Fare showing a wide range due to extreme values. The mean and median values indicate that most passengers paid lower fares, while the presence of outliers increases the average fare.

The histograms show the distribution of numerical variables. The Age variable appears to be slightly right-skewed, with most passengers between 20 and 40 years old. The Fare distribution is highly right-skewed, with most passengers paying lower fares and a few paying very high amounts, indicating the presence of outliers.

The bar charts show the frequency distribution of categorical variables. The dataset contains more male passengers than female passengers. Most passengers traveled in third class, followed by first and second class. The survival distribution indicates that more passengers did not survive compared to those who survived.

The boxplot shows the spread and outliers in the numerical variables. The Fare variable has many extreme outliers, while Age shows moderate variation. SibSp and Parch have low variability and are mostly close to zero.

The pairplot shows relationships between numerical variables. There is no strong linear correlation between most variables, although Fare shows some variation with other features.

The categorical vs categorical analysis shows that female passengers had a much higher survival rate compared to male passengers. This indicates that gender had a strong impact on survival.

The categorical vs numerical analysis shows that survivors tend to be slightly younger, but both groups have overlapping age distributions. This suggests that age alone is not a strong predictor of survival.

The scatter plot of Age vs Fare colored by Survival shows that passengers who paid higher fares had a higher chance of survival. However, there is still overlap between survivors and non-survivors, indicating that multiple factors influenced survival.

```
