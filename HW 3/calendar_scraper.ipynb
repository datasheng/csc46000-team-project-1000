```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.ccny.cuny.edu/registrar/fall'
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html)
print(soup.prettify()[:2000]) # Check if I got the html, I did
```

```python
tables = soup.find_all("table")
rows = tables[0].find_all("tr")
data = []

for row in rows:

    cols = row.find_all("td")
    if len(cols) == 3:
        date = cols[0].get_text(strip=True)
        dow = cols[1].get_text(strip=True)
        text = cols[2].get_text(strip=True)

        data.append((date, dow, text))

data[:5] # Checking if it got what I needed, I did
```

```python
df = pd.DataFrame(data, columns=["date", "dow", "text"])
df.columns = ["Dates", "Day of the Week", "Text with the Explanation"]
df = df.set_index("Dates")

df.head()# Make the dataframe and name them accordingly
```
