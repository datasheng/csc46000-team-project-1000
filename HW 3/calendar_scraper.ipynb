```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.ccny.cuny.edu/registrar/fall'
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html)
print(soup.prettify()[:1000]) # Check if I got the html, I did
```

```python
tables = soup.find_all("table")
rows = tables[0].find_all("tr")
data = []

for row in rows:

    cols = row.find_all("td")
    if len(cols) == 3:
        date = cols[0].get_text(strip=True)
        dow = cols[1].get_text(strip=True)
        text = cols[2].get_text(strip=True)

        data.append((date, dow, text))

data[:5] # Checking if it got what I needed, I did
```

```python
from datetime import datetime

new_data = []

for date, dow, text in data:

    # If the entry is a range like "August 25 - 31"
    date = date.split("-")[0].strip()

    # If the year is already written (January 1, 2022)
    if "2022" in date:
        py_date = datetime.strptime(date, "%B %d, %Y").date()
    else:
        py_date = datetime.strptime(date + " 2021", "%B %d %Y").date()

    # What a pain -_-
    new_data.append((py_date, dow, text))
```

```python
df = pd.DataFrame(new_data, columns=["py_date", "dow", "text"])
df.columns = ["Dates", "Day of the Week", "Text with the Explanation"]
df = df.set_index("Dates")

pd.set_option('display.max_colwidth', 75) # Someone showed me this, cool
df # Make the dataframe and name them accordingly
```
