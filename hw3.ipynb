Scraping the CUNY Fall 2021 academic calendar site

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import io

#import libraries to scrape website
```

```python
url = 'https://www.ccny.cuny.edu/registrar/fall'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
```

```python
df = pd.read_html(io.StringIO(str(soup)), flavor='bs4')
print(df)

#read_html reads any table in the page, when you inspect CUNY Fall 2021 site you can see the information is contained in a table
```

```python
df = df[0] #we must do this because read_html gets a list of all the tables so we only need the first table
df.columns
```

```python
df = df.set_index('DATES')
df.columns
#set the dates as an index
```

```python
df.columns = ['day of the week', 'text with the explanation']
df.columns
#renaming the columns
```
