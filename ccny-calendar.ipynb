```python
import requests

url = "https://www.ccny.cuny.edu/registrar/fall"

response = requests.get(url)

print(response.status_code)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")

print(soup.title)
```

```python
tables = soup.find_all("table")

len(tables)
```

```python
table = tables[0]
rows = table.find_all("tr")

raw = []

for r in rows:
    cols = r.find_all("td")

    if len(cols) >= 3:
        date_text = cols[0].get_text(" ", strip=True)
        dow = cols[1].get_text(" ", strip=True)
        text = cols[2].get_text(" ", strip=True)

        raw.append((date_text, dow, text))

raw[:10]
```

```python
clean = []

for date_text, dow, text in raw:

    dt = pd.to_datetime(date_text + " 2021", errors="coerce")

    if not pd.isna(dt):
        clean.append({
            "date": dt.date(),
            "dow": dow,
            "text": text
        })

df = pd.DataFrame(clean)

df.head()
```

```python
df = df.set_index("date")

df.head(20)
```
